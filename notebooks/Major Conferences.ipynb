```python
import pandas as pd
import numpy as np
```

```python
team_conference = pd.read_csv('../input/kaggle_data/MTeamConferences.csv')
```

```python
def conference(con):
    if con=='pact_ten':
        return 1
    elif con=='pact_twelve':
        return 1
    elif con=='sec':
        return 1
    elif con=='big_eight':
        return 1
    elif con=='big_east':
        return 1
    elif con=='big_twelve':
        return 1
    elif con=='big_east':
        return 1
    elif con=='acc':
        return 1
    else:
        return 0
```

```python
df = team_conference.copy()
df['MajorConf'] = df.apply(lambda x: conference(x.ConfAbbrev),axis=1)

df
```

```python
df.to_csv('../input/major_conferences.csv',index=False)
```
